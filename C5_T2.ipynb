```python
print('COVARIANCE')
```

```python
print('Entire Covariance Matrix (minus ID)')
data = credit.iloc[:,1:25]
covMat = data.cov()
print(covMat)
```

```python
print('Only Covariance of default')
covMat['DEFAULT']
```

```python
print('Plot of Covariance of default')
covDefault = covMat['DEFAULT']
plt.plot(covDefault)
plt.title('Covariance of default')
plt.xlabel('Attribute')
plt.ylabel('Covariance')
```

```python
print('Only Covariance of default (minus credit limit outlier) \n September Bill Amount correlates most negatively \n September Paid Amount is also very negative')
covDefault2 = covDefault.iloc[1:23]
plt.plot(covDefault2)
plt.title('Covariance of default (minus credit limit outlier)')
plt.xlabel('Attribute')
plt.ylabel('Covariance')
```

```python
print('Covariance of default (positive values only) \n September Payment History correlates most positively')
covDefault2 = covDefault.iloc[1:11]
plt.plot(covDefault2)
plt.title('Covariance of default (positive values only)')
plt.xlabel('Attribute')
plt.ylabel('Covariance')
```
